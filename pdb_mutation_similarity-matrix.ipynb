```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(11.7,8.27)})
df_be=pd.read_csv('docking_binding_energy_plot.csv')
df_hb=pd.read_csv('hydrogen_bond_plotdata.csv')
```

```python
df_be.head()
```

```python
df_hb.head()
```

```python
## How we can generate grouped BAR plot in Python


    # loading libraries
import pandas as pd
import matplotlib.pyplot as plt


# Setting the positions and width for the bars
pos = list(range(len(df_hb['Mutant'])))
width = 0.25

# Plotting the bars
fig, ax = plt.subplots(figsize=(20,10))
#fig = plt.figure(figsize =(15,10)) 

# Creating a bar with Maths_score data
plt.bar(pos, df_hb['hbond_native'], width, alpha=0.5, color='#EE3224')
#plt.show()

# Creating a bar with Science_score data,
plt.bar([p + width for p in pos], df_hb['hbond_mutant'], width, alpha=0.5, color='#F78F1E')
#plt.show()

# Creating a bar with French_score data,
plt.bar([p + width*2 for p in pos], df_hb['∆H_bonds'], width, alpha=0.5, color='olive')
#plt.show()

# Setting the y axis label
ax.set_ylabel('No. of hydrogen bonds',fontsize=18)

# Setting the chart's title
ax.set_title('HYDROGEN BOND ANALYSIS ',fontsize=18)

# Setting the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Setting the labels for the x ticks
ax.set_xticklabels(df_hb['Mutant'])
ax.tick_params(axis='both', which='major', labelsize=16)
ax.tick_params(axis='both', which='minor', labelsize=16)
plt.xticks(rotation=60)
# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(df_hb['hbond_native'] + df_hb['hbond_mutant'] + df_hb['∆H_bonds'])] )

# Adding the legend and showing the plot
plt.legend(['hbond_native', 'hbond_native', '∆H_bonds'], loc='upper left',fontsize=18)
plt.grid()
plt. savefig('plot_hb.svg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import plotly.graph_objects as go

plt.rcParams.update({'figure.figsize':(20,15), 'figure.dpi':1000})

fig = go.Figure(go.Waterfall(
    name = "20", orientation = "v",
    measure = ["relative", "relative", "relative","relative", "relative"],
    x = list(df_be['PDB']),
    textposition = "outside",
    text =df_be['Difference'].astype(str) ,
    y =list(df_be['Difference']) ,
    connector = {"line":{"color":"rgb(63, 63, 63)"}},
))

fig.update_layout( title = "∆Binding Energy(B.E_native - B.E_mutant)",
        showlegend = True, scale=2)

fig.show()
```
